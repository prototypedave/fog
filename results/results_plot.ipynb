```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create plots
plt.figure(figsize=(500, 5))
```

```python
# Load data from CSV files
task_completion_times = pd.read_csv('task_completion_time.csv', header=None)

# get the 10th task in the list to create a smooth curve
task_completion = task_completion_times.groupby(0)[1].mean().reset_index()
task_completion = task_completion[::10]
```

```python
# Plot Task Completion Time
plt.plot(task_completion[0], task_completion[1])
plt.xlabel('Number of Tasks')
plt.ylabel('Time (s)')
plt.title('Task Completion Time')
plt.show()
```

```python
# Plot Computational Delay
computation = pd.read_csv('computational_delay.csv', header=None)
computation = computation.groupby(0)[1].mean().reset_index()
```

```python
# Plot Computational Delay
plt.bar(computation[0], computation[1])
plt.xlabel('Node Index')
plt.ylabel('Computation Delay (s)')
plt.title('Computational Delay')
plt.show()
```
